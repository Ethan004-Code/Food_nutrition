```python
import pandas as pd
import os
files = os.listdir()

csv_dfs = {file: pd.read_csv(file) for file in files if file.endswith(".csv")}



for name, df in csv_dfs.items():
    print(f"\n---- {name} ----")
    print(df.head())

```

```python

for name in csv_dfs:
    csv_dfs[name].columns = (
        csv_dfs[name].columns
        .str.lower()
        .str.strip()
        .str.replace(" ", "_")
        .str.replace(r'[^\w\s]', '', regex=True)
    )

for name in csv_dfs:
    df = csv_dfs[name]
    df.dropna(how='all', inplace=True)
    numeric_cols = df.select_dtypes(include='number').columns
    df[numeric_cols] = df[numeric_cols].fillna(0)
    object_cols = df.select_dtypes(include='object').columns
    df[object_cols] = df[object_cols].fillna('Unknown')

    # Try to convert object columns to numeric if possible
    for col in object_cols:
        try:
            df[col] = pd.to_numeric(df[col])
        except:
            continue


for name in csv_dfs:
    before = len(csv_dfs[name])
    csv_dfs[name] = csv_dfs[name].drop_duplicates()
    after = len(csv_dfs[name])
    print(f"{name}: Removed {before - after} duplicate rows")


for name, df in csv_dfs.items():
    print(f"\nCLEANED --- {name} ---")
    print(df.head())
    print(df.info())
```
